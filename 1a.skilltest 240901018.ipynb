```python
import numpy as np

# Exam scores of 20 students
scores = np.array([78, 45, 62, 88, 90, 55, 47, 69, 73, 81,
                   52, 39, 60, 77, 84, 49, 91, 67, 58, 74])

# 1. Overall class performance
average_score = np.mean(scores)
highest_score = np.max(scores)
lowest_score = np.min(scores)

print("Overall Class Performance")
print("Average Score:", average_score)
print("Highest Score:", highest_score)
print("Lowest Score:", lowest_score)

# 2. Pass/Fail classification (pass mark = 50)
pass_mark = 50
passed_students = scores[scores >= pass_mark]
failed_students = scores[scores < pass_mark]

print("\nPass/Fail Classification")
print("Passed Students Scores:", passed_students)
print("Failed Students Scores:", failed_students)

# 3. Grading using predefined thresholds
def assign_grade(score):
    if score >= 85:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'

grades = [assign_grade(score) for score in scores]

print("\nGrades for each student:")
for i, score in enumerate(scores):
    print(f"Student {i+1}: Score = {score}, Grade = {grades[i]}")
```
