```python
import numpy as np

# Impedance values (complex numbers)
Z1 = 10 + 0j      # Resistor
Z2 = 0 - 5j       # Capacitor
Z3 = 0 + 8j       # Inductor

# Store impedances in array
impedances = np.array([Z1, Z2, Z3])

# Ask user for configuration
choice = input("Enter circuit type (series/parallel): ")

if choice.lower() == "series":
    Z_total = np.sum(impedances)

elif choice.lower() == "parallel":
    Z_total = 1 / np.sum(1 / impedances)

else:
    print("Invalid choice")
    Z_total = None

# Display result
if Z_total is not None:
    print("Total Impedance:", Z_total)
```
