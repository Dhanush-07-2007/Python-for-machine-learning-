```python
import numpy as np
import matplotlib.pyplot as plt


grades = np.random.randint(50, 100, size=(30, 3))

subjects = ["Math", "Science", "English"]


avg_scores = np.mean(grades, axis=0)
max_scores = np.max(grades, axis=0)
min_scores = np.min(grades, axis=0)

print("Subject-wise Statistics")
for i in range(3):
    print(f"{subjects[i]} - Avg: {avg_scores[i]}, Max: {max_scores[i]}, Min: {min_scores[i]}")

total_scores = np.sum(grades, axis=1)
avg_student_scores = np.mean(grades, axis=1)


top_student = np.argmax(total_scores)

print("\nTop Performing Student Index:", top_student + 1)
print("Total Score:", total_scores[top_student])

# Visualization (Histogram for each subject)
plt.hist(grades[:,0], alpha=0.5, label='Math')
plt.hist(grades[:,1], alpha=0.5, label='Science')
plt.hist(grades[:,2], alpha=0.5, label='English')

plt.title("Grade Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```
