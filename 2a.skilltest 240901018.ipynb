```python
import numpy as np

# Daily step counts for 30 days
steps = np.array([
    5000, 6200, 7000, 8000, 7600, 9000, 10000,
    8500, 9200, 11000, 12000, 9500, 8700, 7600,
    6800, 7200, 8300, 9100, 10500, 11500,
    9800, 8700, 7600, 6900, 7200, 8100,
    9200, 10000, 10800, 12000
])

# Key statistics
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total Steps in Month:", total_steps)
print("Average Steps per Day:", average_steps)
print("Maximum Steps:", max_steps)
print("Minimum Steps:", min_steps)

# High activity days (threshold = 8000)
threshold = 8000
high_activity_days = np.where(steps > threshold)

print("\nDays with High Activity (steps > 8000):")
print(high_activity_days[0] + 1)  # +1 to show day number

# Trend analysis
trend = np.diff(steps)

print("\nDaily Step Changes:")
print(trend)
```
