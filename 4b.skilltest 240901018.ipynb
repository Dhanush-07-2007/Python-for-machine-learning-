```python
import pandas as pd
import matplotlib.pyplot as plt

# Given dictionary
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price': [1200, 800, 300, 150, 50],
    'Quantity': [10, 20, 15, 25, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Inventory Data:")
print(df)

# Operation 1: Average price
print("\nAverage Price:", df['Price'].mean())

# Operation 2: Maximum price
print("Maximum Price:", df['Price'].max())

# Operation 3: Minimum price
print("Minimum Price:", df['Price'].min())

# Operation 4: Total quantity
print("Total Quantity:", df['Quantity'].sum())

# Operation 5: Add Total Value column
df['Total_Value'] = df['Price'] * df['Quantity']
print("\nData with Total Value:")
print(df)

# Plot bar chart
plt.bar(df['Product'], df['Price'], color='skyblue')
plt.title("Product Price Distribution")
plt.xlabel("Product")
plt.ylabel("Price")
plt.show()
```
