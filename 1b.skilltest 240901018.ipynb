```python
import numpy as np

# Define matrix A (2x2)
A = np.array([[2, 3],
              [1, 2]])

# Define vector b (2x1)
b = np.array([8, 5])

# Solve Ax = b
x = np.linalg.solve(A, b)

# Display result
print("Matrix A:")
print(A)

print("\nVector b:")
print(b)

print("\nSolution vector x:")
print(x)
```
