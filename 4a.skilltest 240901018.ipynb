```python
import pandas as pd
import numpy as np

# Sample EEG dataset (Participant, Electrode readings)
data = {
    "Participant": ["P1","P1","P1","P2","P2","P2","P3","P3","P3"],
    "Electrode": ["E1","E2","E3","E1","E2","E3","E1","E2","E3"],
    "Signal_Value": [12.5, 15.2, 13.8, 14.1, 16.4, 15.0, 13.3, 14.8, 15.6]
}

# Create DataFrame
df = pd.DataFrame(data)

print("EEG Data:")
print(df)

# Group by Participant and Electrode
grouped = df.groupby(["Participant","Electrode"])

# Statistical analysis
mean_values = grouped.mean()
variance_values = grouped.var()
peak_values = grouped.max()

print("\nMean Brainwave Activity:")
print(mean_values)

print("\nVariance of Brainwave Activity:")
print(variance_values)

print("\nPeak Brainwave Activity:")
print(peak_values)
```
